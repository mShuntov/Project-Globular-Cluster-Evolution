```python
#Evolution of stars in the HR diagrams
#Evolution of only one star, here we use M3 data

M3 = loadtxt('M3.dat') 
M1_5 = loadtxt('M1_5.dat')
M15 = loadtxt('M15.dat')
M25 = loadtxt('M25.dat')
M7 = loadtxt('M7.dat')
M85 = loadtxt('M85.dat')



#take only the third column to have ages
age1=M1_5[:,2]
age=M3[:,2]
age2 = M25[:,2]
age3 = M85[:,2]

    
# we can define the size in the following way:
# from the relation: L = 4*pi*R^2*sigma*T^4, the radius would be:
# R ~ sqrt(L/T^4)

ss1_5 = 90 * sqrt(M1_5[:,1]/M1_5[:,0]**4)
ss3 = 90 * sqrt(M3[:,1]/M3[:,0]**4)
ss7 = 90 * sqrt(M7[:,1]/M7[:,0]**4)
ss15 = 90 * sqrt(M15[:,1]/M15[:,0]**4)
ss25 = 90 * sqrt(M25[:,1]/M25[:,0]**4)
ss85 = 90 * sqrt(M85[:,1]/M85[:,0]**4)

# the colour, we correlate with the temperature
cc3 = cm.jet((M3[:,0])/max(M3[:,0]))
cc85 = cm.jet((M85[:,0])/max(M85[:,0]))



for i in range(0,100):

    plot(M3[i,0], M3[i,1],'o-', markersize=ss3[i], label = '$M=3$',color=cc3[i])   #plot the HR diagram one by one
    plot(M1_5[i,0], M1_5[i,1],'o-', markersize=ss1_5[i], label = '$M=1.5$')
    plot(M7[i,0], M7[i,1],'o-', markersize=ss7[i], label = '$M=7$')
    plot(M15[i,0], M15[i,1],'o-', markersize=ss15[i], label = '$M=15$')
    plot(M25[i,0], M25[i,1],'o-', markersize=ss25[i], label = '$M=25$')
    plot(M85[i,0], M85[i,1],'o-', markersize=ss85[i], label = '$M=85$',color=cc85[i])
    
    
    xlim(5,3.50)
    ylim(0,9)
    
    xlabel('log($T_{eff}$)')
    ylabel('log($L$)')
    title('HR Diagram for Zo = 0.001')
    grid()
    legend(loc = 3)

    
    #put the text on the diagram with the age of the star for each step
    figtext(0.21, 0.85, (M3[i,2] * 1.0e-6 ),fontsize = 12)
    figtext(0.13, 0.85, 'AGE:',fontsize = 12, color = 'red',fontweight = 'bold')
    figtext(0.36, 0.85, 'Million Years, for M3',fontsize = 12)
    
    figtext(0.21, 0.8, (M85[i,2] * 1.0e-6 ),fontsize = 12)
    figtext(0.13, 0.8, 'AGE:',fontsize = 12, color = 'red',fontweight = 'bold')
    figtext(0.36, 0.8, 'Million Years for M85',fontsize = 12)
    
    #wait before to plot the next point
    pause(0.02)
    show()
    
    #remove the previous point to have en evolution
    clf()
    
    
#Then we need to generalized this for all stars and fix the problem of the error message after the code running
```
